### Type of customers and Type of travel will never impact the satisfaction of the Airline services so these two columns are not useful for our machine learning model

```python
df_copy=df.drop(['Customer Type','Type of Travel'],axis=1)
```

```python
df_copy.head()
```

## Insights from the above code
1- There are only two columns present in df_copy i.e satisfaction and class which have categorical data

2- Satisfaction column is our dependent feature

3- Remaining columns are our independent features

```python
## Replacement of categorical data with the numerical data in satisfaction column
from sklearn.preprocessing import LabelEncoder
lencoder=LabelEncoder()
df_copy['satisfaction']=lencoder.fit_transform(df[['satisfaction']])
```

```python
df_copy['satisfaction'].unique()
```

### Insights from the above code
1- By using the label encoder we encoded satisfied at 1 and dissatisfied as 0

```python
## Replacement of categorical data with numerical data in class column
from sklearn.preprocessing import OneHotEncoder
ohencoder=OneHotEncoder()
df_copy[['Class_Business','Class_Eco','Class_Eco Plus']]=pd.DataFrame(ohencoder.fit_transform(df_copy[['Class']]).toarray(),
                                                                      columns=[ohencoder.get_feature_names_out()])
```

```python
df_copy[['Class_Business','Class_Eco','Class_Eco Plus']]=df_copy[['Class_Business','Class_Eco','Class_Eco Plus']].astype(int)
```

```python
df_copy.drop('Class',axis=1,inplace=True)
```

```python
df_copy.head()
```

## Insights from the above codes

1- We have been successfully converted all the categorical features into numerical features by using feature scaling techniques

```python
df_copy=df_copy.drop(['Flight Distance','Age'],axis=1)
```

```python
### Independent and dependent features
### Independent features
X=df_copy.drop('satisfaction',axis=1)

### Dependent features
y=df_copy['satisfaction']
```

## Machine Learning Model Implementation

```python
### Train test split for model selection
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
### Standalization of independent variables
from sklearn.preprocessing import StandardScaler
stscaler=StandardScaler()
X_train_stscaled=stscaler.fit_transform(X_train)
X_test_stscaled=stscaler.transform(X_test)
```

```python
### Logistic Regression
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train_stscaled,y_train)
```

```python
## Prediction of dependent variable
y_pred_lg=clf.predict(X_test_stscaled)
```

```python
### Performance Metrix(Confusion metrix,accuracy,classification_Report)
import sklearn.metrics as metrics
print("Accuracy is :", "%.6f" % metrics.accuracy_score(y_test, y_pred_lg))
print("Precision is :", "%.6f" % metrics.precision_score(y_test, y_pred_lg))
print("Recall is :", "%.6f" % metrics.recall_score(y_test, y_pred_lg))
print("F1 Score is :", "%.6f" % metrics.f1_score(y_test, y_pred_lg))
```

```python
### Confusion metrics
cm = metrics.confusion_matrix(y_test, y_pred_lg, labels = clf.classes_)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix = cm,display_labels = clf.classes_)
disp.plot()
```

## Insights from the above code and graphs 

1- For training and testing the data we used logistic regression

2- By logistice regression we got Accuracy, Precision, Recall, and F1 Score is 0.814701, 0.826192, 0.836258, and 0.831194 respectively

3- In confusion metrics we can see that the value of True Positve and True negative is greater than the value of false positive and false negative

4- Difference between the value of false positive and false negative is not huge. 

```python
## Define the parameters for hyperparameter tunning
parameter_lg={'penalty':('l1', 'l2', 'elasticnet', 'None'),
              'solver':('newton-cg', 'sag', 'saga','lbfgs'),
              'max_iter':[100,200]}
```

```python
### Hyperperameter tunning
from sklearn.model_selection import GridSearchCV
gcv_clf_lg=GridSearchCV(clf,param_grid=parameter_lg,scoring='accuracy',cv=5)
gcv_clf_lg.fit(X_train_stscaled,y_train)
```

```python
gcv_clf_lg.best_params_
```

```python
y_pred_gcv_lg=gcv_clf_lg.predict(X_test_stscaled)
```

```python
### Performance Metrix(Confusion metrix,accuracy,classification_Report)
import sklearn.metrics as metrics
print("Accuracy is :", "%.6f" % metrics.accuracy_score(y_test, y_pred_gcv_lg))
print("Precision is :", "%.6f" % metrics.precision_score(y_test, y_pred_gcv_lg))
print("Recall is :", "%.6f" % metrics.recall_score(y_test, y_pred_gcv_lg))
print("F1 Score is :", "%.6f" % metrics.f1_score(y_test, y_pred_gcv_lg))
```

```python
### Confusion metrics
cm = metrics.confusion_matrix(y_test, y_pred_gcv_lg, labels = clf.classes_)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix = cm,display_labels = clf.classes_)
disp.plot()
```
